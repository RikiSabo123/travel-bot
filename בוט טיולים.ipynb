```python
import json
import string
import re
import matplotlib.pyplot as plt
import statistics
from collections import Counter
```

```python
class serviceManager:
    def __init__(self):
        try:
            from dotenv import load_dotenv
            load_dotenv()
            self.client = genai.Client(api_key=os.getenv("GEMINI_API_KEY"))
        except APIError as e:
            print(f"error at connect to API: {e}")
        except Exception as e:
            print(e)    
        self.history=[]
        self.sentiment_dict = {
            # חוויה וטבע
            "מרהיב": 3, "עוצר-נשימה": 3, "פנורמי": 2, "ציורי": 2, "נוף": 2,
            "שקט": 2, "רגוע": 2, "מרענן": 2, "ירוק": 1, "טבעי": 1,
            # אוכל ואירוח
            "טעים": 2, "מעולה": 2, "אותנטי": 2, "טרי": 1, "מגוון": 1,
            "מפנק": 2, "נוח": 1, "נקי": 2, "מומלץ": 2, "שירות": 1,
            # תכנון וארגון
            "מאורגן": 2, "נגיש": 1, "קל": 1, "מהיר": 1, "חסכוני": 2,
            "משתלם": 2, "בטוח": 2, "מדויק": 1, "מקצועי": 2, "מוצלח": 2,
            # ביטויי התלהבות
            "כיף": 2, "מדהים": 3, "נהדר": 2, "מושלם": 3, "חלומי": 3,
            "מרתק": 2, "מיוחד": 2, "שווה": 2, "אהבתי": 2, "מומלץ": 2,
            # שלילי - בעיות תכנון
            "יקר": -2, "מסובך": -1, "מבלבל": -2, "עמוס": -1, "צפוף": -2,
            "מאכזב": -2, "אכזבה": -3, "בעיה": -2, "ביטול": -2, "עיכוב": -1,
            # שלילי - חוויה רעה
            "מלוכלך": -3, "מסוכן": -3, "גרוע": -2, "מייגע": -2, "מתיש": -2,
            "רעש": -1, "מפחיד": -2, "שבור": -2, "סגור": -1, "נכשל": -2
        }
        self.API_connect()
    def API_connect(self):
        chat = self.client.chats.create(
            model="models/gemini-2.5-flash",
            config={"system_instruction": "אתה עוזר טיולים מומחה."}
        )

        response = chat.send_message("מי אתה ומה המומחיות שלך?")
        print(response.text.strip())
        
        self.history = chat.get_history()

        speech = input("את/ה:")
        while speech != "סיום":
            speech = speech.strip()
            
            response = chat.send_message(speech)
            print(response.text.strip())
            
            self.history = chat.get_history()
            
            speech = input("את/ה:")
        self.save_history_as_json(chat.get_history())
        self.draw_graph()

    def save_history_as_json(self, history_objects):
        serializable_history = []
        for msg in history_objects:
            serializable_history.append({"role": msg.role, "parts": msg.parts[0].text})
        
        with open("history.json", 'w', encoding='utf-8') as f:
            json.dump(serializable_history, f, ensure_ascii=False, indent=4)
    def save_2_JSON(self):
        try:
            with open("history.json",'w',encoding='utf-8') as JSON_file:
                json.dump(self.history,JSON_file, ensure_ascii=False, indent=4)
        except IOError as e:
            print(f"error at saving: {e}")
        except Exception as e:
            print(e)
    def draw_graph(self):
        dicti={}
        try:
            with open("history.json",'r', encoding='utf-8') as JSON_file:
                dicti=json.load(JSON_file)
        except IOError as e:
            print(f"error at saving: {e}")
        except Exception as e:
            print(e)
        s_bot=[self.calculate_sentiment(i.get("parts")) for i in dicti if i.get("role")=="model"]
        s_user=[self.calculate_sentiment(i.get("parts")) for i in dicti if i.get("role")=="user"]
        x_bot = range(1, len(s_bot) + 1)
        x_user = range(1, len(s_user) + 1)
        if len(s_bot) > 0:
            average_bot = sum(s_bot) / len(s_bot)
        else:
            average_bot = 0
            
        if len(s_user) > 0:
            average_user = sum(s_user) / len(s_user)
        else:
            average_user = 0
        plt.plot(x_bot, s_bot, label="סנטימנט בוט", marker='o', color='blue', linewidth=2)
        plt.plot(x_user, s_user, label="סנטימנט משתמש", marker='s', color='orange', linewidth=2)
        plt.title("השוואת טון")
        plt.xlabel("מספר הודעה (לפי תפקיד)")
        plt.ylabel("ציון סנטימנט")
        plt.legend()
        plt.show()
        average_bot = sum(s_bot) / len(s_bot)
        average_user = sum(s_user) / len(s_user)
        labels = ['Bot', 'User']
        averages = [average_bot, average_user]
        plt.bar(labels, averages, color=['blue', 'orange'])
        plt.ylabel('Average Sentiment')
        plt.title('Average Sentiment of User vs Bot')
        plt.show()
    def calculate_sentiment(self,text):
        clean_text = "".join([char for char in text if char not in string.punctuation])
        gen_text = (word for word in re.split(r'\s+', clean_text.lower()) if word)
        total_score = sum(self.sentiment_dict.get(word, 0) for word in gen_text)
        return total_score
bot = serviceManager()
```
